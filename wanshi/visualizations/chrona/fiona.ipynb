```python
from pycircos import Garc, Gcircle
import csv
import matplotlib.pyplot as plt
```

```python
#Set chromosomes
circle = Gcircle(figsize=(8,8))
cri_list =[]
fiona_table = "data/fiona-link.csv"
papers = []
breadths = []
evidences = []
color_spec = {"Y":"brown", "P":"grey", "G": "blue", "S": 'red', "H": '#FFFF00', "M": '#CDCD00', "L": '#8B8B00'}
with open(fiona_table) as f:
    first_line = f.readline().rstrip().split(",")
    #print(first_line[1:])
    for item in first_line[1:]:
        papers.append(item)
    breadth = f.readline().rstrip().split(",")
    for item in breadth[1:]:
        #print(item)
        breadths.append(item)
    level_of_evidence = f.readline().rstrip().split(",")
    for item in level_of_evidence[1:]:
        #print(item)
        evidences.append(item)
breadths_dict = dict(zip(papers, breadths))
evidences_dict = dict(zip(papers, evidences))
for title in papers:
    name   = title
    length = 2
    arc    = Garc(arc_id=name, size=length, interspace=2, raxis_range=(935,985), labelposition=80, label_visible=True,label = name[:5], labelsize=10, facecolor=color_spec[evidences_dict[title]], edgecolor=color_spec[breadths_dict[title]], linewidth=1)
    circle.add_garc(arc)

with open(fiona_table) as csvfile:
    for line in csvfile.readlines():
        array = line.split(',')
        first_item = array[0]
        if first_item == 'catagory':
            continue
        #print(first_item)
        length = 2
        arc    = Garc(arc_id=first_item, size=length, interspace=0, raxis_range=(935,985), labelposition=80, label_visible=False)
        circle.add_garc(arc)


# arc  = Garc(arc_id='1', size=10, interspace=0, raxis_range=(835,885), labelposition=80, label_visible=False)
# circle.add_garc(arc)

circle.set_garcs(-90,268)


count = 0
with open(fiona_table) as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        for key, value in row.items():
            if value == 'Y':
                #print((key, 1.5,1.6, 900),(row["catagory"], 0.5, 0.6, 900))
                circle.chord_plot((key, 0.5,0.6, 900),(row["catagory"], 0.5, 0.6, 900), facecolor='red')
            if value == 'P':

                circle.chord_plot((key, 0.55,0.6, 900),(row["catagory"], 0.55, 0.6, 900), facecolor='grey')
        count += 1
        if count >5:
            break
# circle.barplot('CONSORT', data=[1,2,3], raxis_range=[835,885])
for i in papers[:7]:
    circle.fillplot(i, data=[1,2,3], raxis_range=[935,985])
for i in papers[7:18]:
    circle.fillplot(i, data=[3,2,1], raxis_range=[935,985])
for i in papers[18:24]:
    circle.fillplot(i, data=[2,1,2], raxis_range=[935,985])
```

```python
#cytoband
import collections
color_dict   = {"gneg":"#FFFFFF00", "gpos25":"#EEEEEE", "gpos50":"#BBBBBB", "gpos75":"#777777", "gpos100":"#000000", "gvar":"#FFFFFF00", "stalk":"#C01E27", 
               "acen":"#D82322"}

arcdata_dict = collections.defaultdict(dict)
with open("sample_data/example_data_chromosome_cytoband.csv") as f:
    f.readline()
    for line in f:
        line  = line.rstrip().split(",")
        name  = line[0]
        start = int(line[1])-1
        width = int(line[2])-(int(line[1])-1)
        if name not in arcdata_dict:
            arcdata_dict[name]["positions"] = []
            arcdata_dict[name]["widths"]    = []
            arcdata_dict[name]["colors"]    = []
        arcdata_dict[name]["positions"].append(start)
        arcdata_dict[name]["widths"].append(width)
        arcdata_dict[name]["colors"].append(color_dict[line[-1]])

for key in arcdata_dict:
    circle.barplot(key, data=[1]*len(arcdata_dict[key]["positions"]), positions=arcdata_dict[key]["positions"], 
                   width=arcdata_dict[key]["widths"], raxis_range=[935,985], facecolor=arcdata_dict[key]["colors"])    
```

```python
circle.figure
```

```python
#line plot
values_all   = [] 
arcdata_dict = collections.defaultdict(dict)
with open("sample_data/example_data_point.csv") as f:
    f.readline()
    for line in f:
        line  = line.rstrip().split(",")
        name  = line[0]     
        start = int(line[1])-1
        end   = int(line[2]) 
        mid   = (start+end)/2
        value = float(line[-1]) 
        values_all.append(value) 
        if name not in arcdata_dict:
            arcdata_dict[name]["positions"] = []
            arcdata_dict[name]["values"] = []
        arcdata_dict[name]["positions"].append(mid) 
        arcdata_dict[name]["values"].append(value)
    
vmin, vmax = min(values_all), max(values_all) 
for key in arcdata_dict:
    circle.lineplot(key, data=arcdata_dict[key]["values"], positions=arcdata_dict[key]["positions"], 
                    rlim=[vmin-0.05*abs(vmin), vmax+0.05*abs(vmax)], raxis_range=(755,835), linecolor="royalblue", spine=False)
```
